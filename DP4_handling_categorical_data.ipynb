# ordinal Number Encoding

```python
import datetime
import pandas as pd
```

```python
today_date = datetime.datetime.today()
today_date
```

```python
today_date - datetime.timedelta(3)
```

```python
#calculate first 15 days
days =[today_date - datetime.timedelta(x) for x in range(0,15)]
```

```python
#convert days into in dataframe
data = pd.DataFrame(days)
data.columns=['Day']
data.head(10)
```

```python
data['Weekday'] = data['Day'].dt.day_name()
data.head()
```

```python
dict= {'Sunday':1,
      'Monday':2,
      'Tuesday':3,
      'Wednesday':4,
      'Thursday':5,
      'Friday':6,
      'Saturday':7
      }
dict
```

```python
data['weekday_ordinal'] = data['Weekday'].map(dict)
data
```

# Target Guided oridinal Encoding

```python
df = pd.read_csv('titanic.csv',usecols = ['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace =True)
df
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0] #astype means a string types
```

```python
df
```

```python
df.Cabin.unique()
```

```python
ordinal_label = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_label
```

```python
enumerate(ordinal_label)
```

```python
ordinal_values = {k:i for i,k in enumerate(ordinal_label,0)}
```

```python
df['Cabin_ordinal_values']  = df['Cabin'].map(ordinal_values)
df.head()
```

# Mean Encoding

```python
mean_ordinal = df.groupby(['Cabin'])['Survived'].mean().to_dict()
mean_ordinal
```

```python
df['mean_encoding'] = df['Cabin'].map(mean_ordinal)
df.head()
```

# probality ratio Encoding

1. Probability of Survived based on Cabin--- Categorical Feature
2. Probability of Not Survived---1-pr(Survived)
3. pr(Survived)/pr(Not Survived)
4. Dictonary to map cabin with probability
5. replace with the categorical feature

```python
prob_df = pd.read_csv('titanic.csv',usecols = ['Survived','Cabin'])
prob_df
```

```python
prob_df['Cabin'].fillna('Missing',inplace =True)
prob_df.head()
```

```python
prob_df['Cabin'].unique()
```

```python
len(prob_df['Cabin'].unique())
```

```python
prob_df['Cabin']= df['Cabin'].astype(str).str[0]
prob_df.head(10)
```

```python
pro_df = prob_df.groupby(['Cabin'])['Survived'].mean()
pro_df
```

```python
pro_df = pd.DataFrame(pro_df)
pro_df
```

```python
pro_df['Died'] = 1- pro_df['Survived']
pro_df
```

```python
pro_df['probality_ratio'] = pro_df['Survived']/pro_df['Died']
pro_df
```

```python
probality_encoding = pro_df['probality_ratio'].to_dict()
probality_encoding
```

```python
prob_df['cabin_encoding'] = prob_df['Cabin'].map(probality_encoding)
prob_df.head(20)
```
