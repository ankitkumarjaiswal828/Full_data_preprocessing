####  5th method :-  Arbitary values imputation :- it is consist of replacing nan values by arbitary values.This technique drived by kaggle compitation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
arbi_df = pd.read_csv('titanic.csv',usecols= ['Age','Fare','Survived'])
arbi_df.head(10)
```

```python
arbi_df.isnull().sum()
```

```python
def impute_nan(arbi_df,variable):
    arbi_df[variable+'_Zero'] = arbi_df[variable].fillna(0)
    arbi_df[variable+'_Hundred'] = arbi_df[variable].fillna(100)
```

```python
impute_nan(arbi_df,'Age')
arbi_df.head(25)
```

histogram is used to show outlier of data like less outlier on 80 on x axis and high outlier on 20 x axis

```python
arbi_df['Age'].hist(bins=50)
```

Advantages:-
1. capturing the importance of missing values 

disadvantages:-
1. Distorts the original distribution of the variable
2. If missingess is not important, it may mask the predictive power of the original variable by distorting its distribution
3. Hard to decide which value to use

```python
def impute_nan(c_df,variable):
    most_frequent = c_df[variable].mode()[0]
    c_df[variable].fillna(most_frequent,inplace =True)
    
for features in ['Age','Cabin','Embarked']:
    impute_nan(c_df,features)
```

```python
c_df.isnull().sum()
```

```python
c_df.isnull().mean()
```

disadvantage :-
1. It distorts the relation of most frequent label
2. Since we are using the more frequent labels, it may use them in an over respresented way, if there are many nan's

# Adding a variable to capture NaN

```python
df2 = pd.read_csv('titanic.csv',usecols =['Age','Cabin','Embarked'])
df2.isnull().sum()
```

```python
df2['Age_var'] = np.where(df2['Age'].isnull(),1,0)
df2.head(20)
```

```python
def impute_nan(df2,variable):
    df2[variable+'_var'] = np.where(df2[variable].isnull(),1,0)

for i in ['Age','Cabin','Embarked']:
    impute_nan(df2,i)
```

```python
df2.head(20)
```

```python
df2['Age'].mode()[0]
```

Suppose if you have more frequent categories, we just replace NAN with a new category

```python
def missing_val(df2,variable):
    df2[variable+'_nan_missing'] = np.where(df2[variable].isnull(),'Missing',df2[variable])
    
for i in ['Age','Cabin','Embarked']:
    missing_val(df2,i)
```

```python
df2.head(20)
```

```python
df2.drop(columns = ['Age_var','Cabin_var','Embarked_var'],inplace =True)
```
