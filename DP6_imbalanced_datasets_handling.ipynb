# cross validation like KFold and Hyperparameter Tuning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import KFold
```

parameter and hyperparameter tuning is used to increase performance and accuracy of algorithms and models

```python
num = 10.0**np.arange(-2,3)
num
```

```python
from sklearn.model_selection import GridSearchCV
log_class = LogisticRegression()
grid = {'C':num,'penalty':['l1','l2']}
cv = KFold(n_splits =5,random_state = None,shuffle =False)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size = 0.7,test_size =0.3)
```

```python
clf = GridSearchCV(log_class,grid,cv = cv,n_jobs = -1 , scoring = 'f1_macro')
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

# Under Sampling Technique

```python
from collections import Counter
Counter(y_train)
```

```python
#first method for finding the resampling of x,y train data
from imblearn.under_sampling import RandomUnderSampler  
under_sampler = RandomUnderSampler()
x_train_res, y_train_res = under_sampler.fit_resample(x_train, y_train)
print('The number of classes before fit {}'.format(Counter(y_train)))
print('The number of classes after fit {}'.format(Counter(y_train_res)))
```

```python
#second method for finding the resampling of x,y train data
from imblearn.under_sampling import NearMiss
ns = NearMiss(0.8)
x_train_ns,y_train_ns = ns.fit_resample(x_train,y_train)
print('The number of classes before fit {}'.format(Counter(y_train)))
print('The number of classes after fit {}'.format(Counter(y_train_ns)))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(x_train_res,y_train_res)
```

```python
y_pred = classifier.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# over Sampling technique

```python
from imblearn.over_sampling import RandomOverSampler

os = RandomOverSampler(0.75)
x_train_res,y_train_res = os.fit_resample(x_train,y_train)
print('The Number of classes before fit {}'.format(Counter(y_train)))
print('The number of classes after resampler/fit {}'.format(Counter(y_train_res)))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(x_train_res,y_train_res)
```

```python
y_pred = classifier.predict(x_test)
print(confusion_matrix(y_test,y_pred))

print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
